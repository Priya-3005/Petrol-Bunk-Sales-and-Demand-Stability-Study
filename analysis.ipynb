# DEMAND STABILITY INDEX 

### 
The Demand Stability Index (DSI) is a custom metric designed to measure the consistency of fuel demand over time. It is based on the coefficient of variation, which captures the relative dispersion in sales data.
* If demand variation is high - low DSI (unstable)
* If demand variation is low  - high DSI (highly stable) 
* Coeffecient of variation (CV)
    * CV= Standard Deviation/ Mean
* DSI 
    * DSI = 1 / (1+CV) 
* This formulation ensures that lower variability in sales leads to a higher DSI value, indicating stable demand, while higher variability results in a lower DSI, reflecting instability.
- If variation is small → CV is low → DSI → close to 1 (stable)
- If variation is high → CV is high → DSI → closer to 0 (unstable)

```python
mean_sales = petrol_data['Total_Fuel_Sales'].mean()
std_sales = petrol_data['Total_Fuel_Sales'].std()

cv = std_sales / mean_sales

dsi = 1 / (1 + cv)

print("Demand Stability Index (DSI):", dsi)
```

##### The calculated Demand Stability Index (DSI) for the dataset is 0.853, indicating a high level of demand stability. This suggests that fuel sales exhibit minimal fluctuations over time, making demand predictable and consistent. Such stability is beneficial for inventory planning, operational efficiency, and revenue forecasting.

##### Station wise DSI 

```python
station_dsi = petrol_data.groupby('Station_Name')['Total_Fuel_Sales'].agg(['mean','std'])
station_dsi['CV'] = station_dsi['std'] / station_dsi['mean']
station_dsi['DSI'] = 1 / (1 + station_dsi['CV'])

print(station_dsi)
```

##### A station-wise analysis of the Demand Stability Index reveals that all stations exhibit high stability in fuel demand. Among them, Iwaya station shows the highest stability (DSI = 0.857), followed closely by Idimu (DSI = 0.854), while Maryland has a slightly lower but still strong stability score (DSI = 0.849). The minimal variation across stations indicates uniform demand patterns and consistent customer behavior across locations.
